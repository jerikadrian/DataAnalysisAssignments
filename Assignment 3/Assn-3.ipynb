### Figure 2

Use `data-fig2.npy` for this figure.

```python
a = np.load( 'data-fig2.npy' )
x = np.arange(a.shape[0])

colors = ['red', 'green', 'blue']
labels = ['Red', 'Green', 'Blue']

for i in range(a.shape[1]):
    plt.plot(x, a[:, i], color=colors[i], label=labels[i])

plt.xlabel('X value')
plt.ylabel('Y value')
plt.legend()
plt.show()
```

### Figure 3

Use `data-fig3.npy` for this figure.

```python
a = np.load( 'data-fig3.npy' )

x = a[:, 0]
y = a[:, 1]

plt.scatter(x, y, color='blue', s=200, alpha=0.3)

plt.xlabel('X value')
plt.ylabel('Y value')

plt.show()
```

### Figure 4

Use `data-fig4.npy` for this figure.

```python
a = np.load( 'data-fig4.npy' )

import string  # gives us AZ letters

labels = list(string.ascii_uppercase[:len(a)])

plt.bar(labels, a, color='lightgray')  # you can change color

plt.xlabel('Category')
plt.ylabel('Value')

plt.show()
```

### Figure 5

Use `data-fig5.npy` for this figure.

*Hint*: Use 2D visualization and the "RdBu" colormap.

```python
a = np.load( 'data-fig5.npy' )

plt.imshow(a, cmap='RdBu', origin='lower', vmin=-1, vmax=1)
plt.colorbar()

plt.gca().set_aspect(33/100)  

plt.xlim(0, 40)   
plt.ylim(0, 120) 

plt.xticks(np.arange(0, 41, 5))  

plt.show()
```
