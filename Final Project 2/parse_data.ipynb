# parse_data.ipynb

This notebook parses the data files used for the FP-2 assignment. 

<br>
<br>

First let's read the attached data file:

```python
import pandas as pd

df0 = pd.read_csv('energydata_complete.csv')

df0.describe()
```

<br>
<br>

The dependent and independent variables variables (DVs and IVs) that we are interested in are:

**DVs**:
- None since I am planning to do a clustering analysis

**IVs**:
- Energy Consumption from Aplliances (Wh) ("Appliances" column in the CSV file)
- Energy Consumption from Lights (Wh) ("lights" column in the CSV file)
- Temperature in the Living Room (C)  ("T1" column in the CSV file)
- Humidity in the Living Room (%)  ("RH_1" column in the CSV file)
- Temperature Outdoors (C)  ("T_out" column in the CSV file)
- Humidity Outdoors (%) ("RH_out" column in the CSV file)


<br>
<br>

Let's extract the relevant columns:

```python

df = df0[  ['Appliances', 'lights', 'T1', 'RH_1', 'T_out', 'RH_out']  ]

df.describe()
```

<br>
<br>

Next let's use the `rename` function to give the columns simpler variable names:

```python
df = df.rename( columns={'Appliances':'app', 'T1':'T_in', 'RH_1':'RH_in'} )

df.describe()
```
