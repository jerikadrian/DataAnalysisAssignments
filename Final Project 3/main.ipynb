### Purpose:

* The purpose of this Final Project is to determine the relationship of indoor–outdoor environmental conditions that contribute to the energy consumption of a low-energy household.
* The key **dependent variable** (DV) is total energy consumption from appliances and lights.
* Key **independent variables** (IVs) include:
    * temperature indoors
    * temperature outdoors
    * humidity indoors
    * humidity outdoors
    * windspeed
    * visibility
* This dataset contains 19735 cases because it is recorded every 10 minutes, but I aggregated it in order for the data to be shown daily. I got the sum of the energy consumed for appliances and lights, while I got the mean of the different environmental conditions.
